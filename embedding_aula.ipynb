```python
# é criada a coleção que será utilizada
client.create_collection(
    collection_name="items", #a coelação criada se chama "Items"
    vectors_config=VectorParams(size=384, distance=Distance.COSINE), #é estabelecida a configuração dos vetores, o tamanho é de 348 e a similariedade é dada pela distancia dos cossenos
)
```

```python
# Gerar a consulta para embedding, no caso em questão a consulta é para retornar pratos vegetarianos
query_text = "vegetarian dishes"

# Busca para encontrar os itens que são similares a questão de consulta
results = client.query_points(
    collection_name="items",
    query=Document(text=query_text, model="sentence-transformers/all-MiniLM-L6-v2"),
    with_payload=True,
    limit=5
)

# Estrutura para printar os resultados
for result in results.points:
    print(f"Item: {result.payload.get('item_name', 'N/A')}")
    print(f"Score: {result.score}")
    print(f"Description: {result.payload['description'][:150]}...")
    print(f"Price: {result.payload.get('price', 'N/A')}")
    print("---")
```
